```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import pandas as pd
import altair as alt
```

```python
data = pd.read_csv("data/new_return_data.csv")
data['visit_date'] = pd.to_datetime(data['visit_date'])
data = data.set_index('visit_date')
data
```

```python
rolling_data = data.drop(['newUsers','returningUsers'],axis='columns')
rolling_data['New'] = rolling_data['new_users']
rolling_data['Returning'] = rolling_data['return_users']
rolling_data = rolling_data.drop(['new_users','return_users'],axis='columns')
rolling_data = rolling_data.loc[:, ~rolling_data.columns.str.contains('^Unnamed')]
rolling_data = rolling_data.unstack().reset_index() 
rolling_data['visit_date'] = pd.to_datetime(rolling_data['visit_date'])
#rolling_data = rolling_data.set_index('visit_date')
rolling_data = rolling_data.rename(columns={"level_0": "User_type", 0: "total_users"})
```

```python
rolling_data
```

```python
chart = alt.Chart(rolling_data).mark_area().encode(
    alt.X('visit_date:T',title='Date'),
    alt.Y('sum(total_users):Q', stack='center', axis = None, title='Total Visits by user type'),
    alt.Color('User_type:N',
        scale=alt.Scale(scheme='category20'),
              title='User Category'
    )
).properties(
    width=600,
    height=500,
    title = 'New and returning users over 1 year'
).interactive()

chart.configure_header(
    titleColor='grey',
    titleFontSize=22,
)
```
